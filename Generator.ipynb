Let's Create our own generator to generate the cubes of number.

```python
def cube(n):
    result = []
    for x in range(0,n):
        result.append(x**3)     ## this is not a efficient way of doing it as it will create a list everytime ,taking unwanted memory .
    return result
```

```python
for _ in cube(10):
    print(_)
```

Remember RANGE is a Generator which generates number rather storing it to list .
if we somehow need the list we just convert it to list ------> list(range(0,n))

```python
## Creating cube generator !!
def cubes(n):

    for x in range(0,n):
        yield x**3
```

```python
cubes(10)     # it will not return a list .
```

```python
for _ in cubes(10):
    print (_)
```

```python
list(cubes(10))     # if we need the list !
```

```python
## Generating fibbonacci sequence of numbers .
def gen_fibo(n):
    a=1
    b=1
    for x in range(n):
        yield a
        (a,b) = (b,a+b)
```

```python
for _ in gen_fibo(10):
    print(_)
```

```python
def simple_gen():
    for x in range(0,3):
        yield x
```

```python
simple_gen()
```

```python
for _ in simple_gen():
    print (_)
```

```python
g = simple_gen()
```

Iter function 

```python
s = "Hello"
```

```python
for _ in s :
    print (_)
```

```python
next(s)    # not an iterable object !
```

```python
s_iter = iter(s)    # converts into iterable object !
```

```python
next(s_iter)
```
