We can return a function within another function !

```python
def hello(name="Rohit"):
    print("Hey ! This is hello() function .")

    def greet():
        return ("\t this is greet() function inside hello()")
    def welcome():
        return ("\t this is welcome() function inside hello()")
    print("\t I am going to return a function !")

    if name == "Rohit":
        return greet
    else:
        return welcome 
```

```python
my_func = hello()
```

```python
my_func     
```


.

Currently my_func is pointing towards hello() function which is returning the greet() function .

so we can now execute my_func() which will be returning greet() function .

```python
my_func()       #returning greet()
```

```python
print(my_func())     #printing greet()
```

```python
my_func_1 = hello("Jossee")
```

```python
my_func_1         #pointing towards welcome()
```

```python
my_func_1()       #returning welcome()
```

```python
print(my_func_1())        #printing out welcome()
```

Let's create a decorator function !! which add functionalities to the original function .

```python
def decorator_func(original):
    def wrap():
        print("Extra code before the original function !")
        original()
        print("Extra code after the original function !")
    return wrap

## ADD FUNCTIONALITIES HERE !!!!
```

```python
def original_func():
    print(" \t I am the original function ")    # original function !
```

```python
decorated_func = decorator_func(original_func)
```

```python
decorated_func()
```

```python
## in one line 
@decorator_func
def original_func():
    print(" \t I am the original function ")    # original function !
```

```python
original_func()
```
