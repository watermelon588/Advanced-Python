```python
file_path = 'C:\\Users\\Rohit Maity\\Desktop\\coding\\Python coding\\Exp_top_level'
```

```python
for (folder,sub_folder,files) in os.walk(file_path):
    print(f"Currently looking at {folder}")
    print("\n")
    print("Sub Folders are : ")
    for sub_fold in sub_folder:
        print(f"\t subfolders : {sub_fold}")
    print("\n")
    print("Files are : ")
    for file in files :
        print(f"\t Files : {file}")
    print("\n")
```
